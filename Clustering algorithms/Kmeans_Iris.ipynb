**Application A : Iris dataset**

Ce jeu de données, qui a été utilisé pour illustrer le fonctionnement d'un grand nombre d'algorithme a entre autres une description sur [wikipedia](https://en.wikipedia.org/wiki/Iris_flower_data_set).    
Le jeu de données comprend 50 échantillons de chacune des **trois espèces d'iris** (`Iris setosa`, `Iris virginica` et `Iris versicolor`).  **Quatre caractéristiques ont été mesurées** à partir de chaque échantillon : **la longueur et la largeur des sépales et des pétales**, en centimètres. Sur la base de la combinaison de ces quatre variables, [Fisher](https://fr.wikipedia.org/wiki/Ronald_Aylmer_Fisher) a élaboré un modèle linéaire permettant de distinguer les espèces les unes des autres. 

```python
%matplotlib inline
import numpy as np
import matplotlib 
import matplotlib.pyplot as plt 
from sklearn import datasets
from sklearn.manifold import TSNE
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_samples, silhouette_score, adjusted_rand_score
from sklearn.preprocessing import StandardScaler
import pandas as pd
#from sklearn_extra.cluster import KMedoids
from scipy.spatial import distance
#import pandas_profiling as pp
```

```python
def compute_bic(kmeans,X):
    """
    Computes the BIC metric for a given clusters
    Parameters:
    -----------------------------------------
    kmeans:  List of clustering object from scikit learn
    X     :  multidimension np array of data points
    Returns:
    -----------------------------------------
    BIC value
    """
    # assign centers and labels
    centers = [kmeans.cluster_centers_]
    labels  = kmeans.labels_
    #number of clusters
    m = kmeans.n_clusters
    # size of the clusters
    n = np.bincount(labels)
    #size of data set
    N, d = X.shape
    #compute variance for all clusters beforehand
    cl_var = (1.0 / (N - m) / d) * sum([sum(distance.cdist(X[np.where(labels == i)], [centers[0][i]], 'euclidean')**2) for i in range(m)])
    const_term = 0.5 * m * np.log(N) * (d+1)
    BIC = np.sum([n[i] * np.log(n[i]) -
               n[i] * np.log(N) -
             ((n[i] * d) / 2) * np.log(2*np.pi*cl_var) -
             ((n[i] - 1) * d/ 2) for i in range(m)]) - const_term
    return(BIC)
```

```python
# load dataset
(data,target) = datasets.load_iris(return_X_y=True)
nomDesVariables = ['hauteur sépale','largeur sépale','hauteur pétale','largeur pétale']
df_iris = pd.DataFrame(data,columns=nomDesVariables)
df_iris.describe()
```

**Visualisation de données**

Afin de pouvoir visualiser la bae de données, nous en allons projeter dans un plan 2D en utilisant la décomposition ACP

```python
pca = PCA(n_components=2)
pca.fit(data)
X_pca = pca.transform(data)
labels = np.unique(target)
color=['r','g','b']

plt.figure(figsize=(10, 6))
for l in labels:
    plt.plot(X_pca[np.where(target==l),0],X_pca[np.where(target==l),1], color[l]+'.')
#plt.legend()    
```

**Vérification de nombre de clusters**

Pour cela, nous allons utiliser les 3 méthodes qu'on a déja cité dans la partie introduction :
- L'inertie intra-classe
- Le score BIC
- Le score moyen de Silhouette

```python
# liste_n_clusters donne le nbre de clusters considérés dans chaque expérience
X = data
#X = StandardScaler().fit_transform(data) 
X = X_pca

liste_n_clusters=[1, 2, 3, 4, 5, 8, 10, 15, 20,  len(X)//2, len(X)]
liste_n_clusters
```

```python
# les paramétres
rand = list()
inerties_intra = list()
bic_clusters = list()
silhouette_avg = list()
silhouette_sample_values = list()

for i,n_clusters in enumerate(liste_n_clusters):
    
    # création du modèle de classification
    k_means=KMeans(n_clusters=n_clusters)
    k_means.fit(X)

    # calcul de l'inertie intra classes
    if n_clusters==1:
        inertie_totale= k_means.inertia_
    inertie_intra_=100 * k_means.inertia_ / inertie_totale
    inerties_intra.append(k_means.inertia_)

    # Calcul du BIC de la partition
    if n_clusters!=1 and n_clusters!=X.shape[0]:
        bic_clusters.append(compute_bic(k_means,X))

    # Calcul du coefficient de silhouette 
    #  (silhouette_score gives the average value for all the samples)
    if n_clusters!=1 and n_clusters!=X.shape[0]:
        silhouette_avg.append(silhouette_score(X, k_means.labels_))
        silhouette_sample_values.append(silhouette_samples(X, k_means.labels_))
    
    # calcul rand
    rand.append(adjusted_rand_score(k_means.labels_,target))
```

```python
plt.figure(figsize=(18,6))
plt.plot(liste_n_clusters[:-3],inerties_intra[:-3],'o-', markersize = 8)
plt.grid()
plt.xlabel('n_clusters (k)',fontsize=18)
plt.ylabel('Inertie intra-classe',fontsize=18)
```

```python
plt.figure(figsize=(18,6))
plt.plot(liste_n_clusters[1:-3],bic_clusters[:-2],'o-') # lors de calcul de BIC : on exclut le 1 et len(Data)
plt.grid()
plt.xlabel('n_clusters (k)',fontsize=18)
plt.ylabel('BIC',fontsize=18) ;
```

```python
plt.figure(figsize=(18,6))
plt.plot(liste_n_clusters[1:-3],silhouette_avg[:-2],'o-')  # lors de calcul de Silh : on exclut le 1 et len(Data)
plt.grid()
plt.xlabel('''nombre d'aggrégats (k)''',fontsize=18)
plt.ylabel('coefficient de silhouette',fontsize=18)
```

**Analyse des résultats**

- Pour la méthode intra-classe, nous constatons un coude d'inertie pour k = 2. Neanmoins, le coefficient BIC nous montre que le nombre de cluster n'est pas judicieux. 
- Par rapport au coef Bic, nous voyons le score est maximal pour k = 5, mais l'ecart de BIC entre k = 3 et k = 5 n'est pas trop intéressant de même pour l'inertie intra-classe. Pour ce fait, si on a pas une idée claire que le nombre de clusters, nous allons considerer la valeur la plus petite et valider lors de visualisation es frontières de décision

#### Frontières de Décision

```python
# Paramètres
n_classes = 3
clf = KMeans(n_clusters=n_classes).fit(X)
```

```python
## Affichage de la surface de décision
plot_step = 0.2
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, plot_step), np.arange(y_min, y_max, plot_step))

# prediction label
Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

# plotting 
plt.figure(figsize=(8, 8))
cs = plt.contourf(xx, yy, Z, cmap=plt.cm.Paired)
plt.xlabel("axis 1")
plt.ylabel("axis 2")

for l in labels:
    plt.plot(X_pca[np.where(target==l),0],X_pca[np.where(target==l),1], color[l]+'.')
    
plt.plot(clf.cluster_centers_[:, 0],clf.cluster_centers_[:,1], 'yD',
         markersize=12, label = "Cluster centres")

plt.axis("tight")
plt.suptitle("Decision surface")
plt.legend()
plt.show()
```
